```python
# Create a list of coordinates from the df, with x being the loss_improvement and y being the perplexity:
x = df['loss_improvement']
y = df['perplexity']

# Convert to list:
x = x.tolist()
y = [float(num[7:-1]) for num in y.tolist()]

print(x[:5])
print(y[:5])

# Plot the coordinates:
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.show()
```
